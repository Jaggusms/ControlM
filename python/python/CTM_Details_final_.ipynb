```python
m=input("enter the name of xml file:")
e=input("enter the environment keyword:")
#this is job code
import xml.etree.ElementTree as et
parsexml=et.parse(m+'.xml') 
root=parsexml.getroot()
#The header do you want
keypoint=['JOBNAME','PARENT_FOLDER','APPL_TYPE','DESCRIPTION','TASKTYPE','CMDLINE','NODEID','RUN_AS','APPLICATION','SUB_APPLICATION','PRIORITY','CONFIRM','RETRO','DAYS','JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC','TIMEFROM','TIMETO','CYCLIC','MAXRERUN','INTERVAL','MAXWAIT']
vaa=[]# data of the headers
for r in root:       #getting root tag attribues in 
  JOB=r.findall('JOB')   #finding the all attribuets of job
  for j in JOB:
      k=list(j.attrib.keys())       
      v=list(j.attrib.values())
      va=[]
      for q in keypoint[:32]:  #30 is length of header as for the first
          if q in k:
            va.append(v[k.index(q)])
          else:
            va.append('')
      # variable keyword details appending in the data
      VARIABLE=j.findall('VARIABLE')
      L_VARIABLE=[]
      if VARIABLE:
        if 'VARIABLE' not in keypoint:
          keypoint.append('VARIABLE')
        st1=''
        for B in VARIABLE:
          s=list(B.attrib.values())
          st1 +=str(s[0])+' '+str(s[1])+'  '     
        L_VARIABLE.append(st1)
        va.append(L_VARIABLE)
      else:
        if 'VARIABLE' not in keypoint:
          keypoint.append('VARIABLE')
        va.append('')

      #RULE BASED CALENDERS details appending in the data  
      RBC=j.findall('RULE_BASED_CALENDARS')
      L_RBC=[]
      if RBC:
        if 'RBC' not in keypoint:
          keypoint.append('RBC')
        st=''
        for a in RBC:
          s=list(a.attrib.values())
          st +=str(s[0])+'  '
        L_RBC.append(st)
        va.append(L_RBC)
      else:
        if 'RBC' not in keypoint:
          keypoint.append('RBC')
        va.append('')
      
      #CONTROL details appending in the data
      CONTROL=j.findall('CONTROL')
      L_CONTROL=[]
      if CONTROL:
        if 'CONTROL' not in keypoint:
          keypoint.append('CONTROL')
        st=''
        for a in CONTROL:
          s=list(a.attrib.values())
          st +=str(s[0])+' '+str(s[1])+'   '
        L_CONTROL.append(st)
        va.append(L_CONTROL)
      else:
        if 'CONTROL' not in keypoint:
          keypoint.append('CONTROL')
        va.append('')
      
      #QUANTITATIVE details appending in the data
      QUA=j.findall('QUANTITATIVE')
      L_QUA=[]
      if QUA:
        if 'QUANTITATIVE' not in keypoint:
          keypoint.append('QUANTITATIVE')
        st=''
        for a in QUA:
          s=list(a.attrib.values())
          st +=str(s[0])+' '+str(s[1])+'   '
        L_QUA.append(st)
        va.append(L_QUA)
      else:
        if 'QUANTITATIVE' not in keypoint:
          keypoint.append('QUANTITATIVE')
        va.append('')
      
      # Inconditions appending to the data
      INCOND=j.findall('INCOND')
      L_INCOND=[]
      if INCOND:
        if 'INCOND' not in keypoint:
          keypoint.append('INCOND')
        st=''
        for a in INCOND:
          s=list(a.attrib.values())
          st +=str(s[0])+' '+str(s[1])+" "+str(s[2])+' '
        L_INCOND.append(st)
        va.append(L_INCOND)
      else:
        if 'INCOND' not in keypoint:
          keypoint.append('INCOND')
        va.append('')
      
      # Outcondition appending to the data
      OUTCOND=j.findall('OUTCOND')
      L_OUT=[]
      if OUTCOND:
        if 'OUTCOND' not in keypoint:
          keypoint.append('OUTCOND')
        st=''
        for a in OUTCOND:
          s=list(a.attrib.values())
          st +=str(s[0])+' '+str(s[1])+" "+str(s[2])+' '
        L_OUT.append(st)
        va.append(L_OUT) 
      else:
        if 'OUTCOND' not in keypoint:
          keypoint.append('OUTCOND')
        va.append('')

      # On conditin appending to the data
      ON=j.findall('ON')
      L_ON=[]
      if ON:
        if 'ON' not in keypoint:
          keypoint.append('ON')
        key=['CODE']
        st=''
        for p in ON:
          k=list(p.attrib.keys())
          v=list(p.attrib.values())
          for q in key:
            if q in k:
              st +=str(v[k.index(q)])+'  '
            DOACTION=p.findall('DOACTION')
            if DOACTION:
              for aB in DOACTION:
                l=list(aB.attrib.values())
                st +=" ".join(l)+'  '
            DOCOND=p.findall('DOCOND')
            if DOCOND:
              for aB in DOCOND:
                l=list(aB.attrib.values())
                st +=" ".join(l)+'  '
            DOEMAIL=p.findall('DOMAIL')
            if DOEMAIL:
              for aB in DOEMAIL:
                l=list(aB.attrib.values())
                st +=" ".join(l)+'  '  
            DOSHOUT=p.findall('DOSHOUT')
            if DOSHOUT:
              for aB in DOSHOUT:
                l=list(aB.attrib.values())
                st +=" ".join(l)+'  '  
        L_ON.append(st)
        va.append(L_ON)           
      else:
        if 'ON' not in keypoint:
          keypoint.append('ON')
        va.append('')
      SHOUT=j.findall('SHOUT')
      L_SHOUT=[]
      if SHOUT:
        if 'NOTFY BEFOR(0)/AFTER' not in keypoint:
          keypoint.append('NOTFY BEFOR(0)/AFTER')
        st=''
        for a in SHOUT:
          s=list(a.attrib.values())
          st +=" ".join(s)+'  '
        L_SHOUT.append(st)
        va.append(L_SHOUT) 
      else:
        if 'NOTFY BEFOR(0)/AFTER' not in keypoint:
          keypoint.append('NOTFY BEFOR(0)/AFTER')
        va.append('')  
      #final data append to to vaa list
      vaa.append(va)
#import the csv module to store the data
import csv
with open(m+"_"+e+"_JOB.csv", "w", newline='') as f:    #file name that you want to save only in CSV
    writer = csv.writer(f, delimiter=',')
    writer.writerow(keypoint) 
    for l in vaa:
        writer.writerow(l) 
print("succuss_JOB")



#actuval folder code
import xml.etree.ElementTree as et
parsexml=et.parse(m+'.xml')
root=parsexml.getroot()
keypoint=['TASKTYPE','FOLDER_NAME','DESCRIPTION','DATACENTER','FOLDER_ORDER_METHOD','RUN_AS','APPLICATION','SUB_APPLICATION','SITE_STANDARD_NAME','TIMEFROM','TIMETO','CYCLIC','INTERVAL','MAXWAIT','MAXRERUN']
vaa=[]
for r in root:
  k=list(r.attrib.keys())
  v=list(r.attrib.values())
  va=[]
  for q in keypoint[:15]:
    if q in k:
      va.append(v[k.index(q)])
    else:
      if q=='FOLDER_ORDER_METHOD':
        va.append('None Manuval Order')
      else:
        va.append('')
  VARIABLE=r.findall('VARIABLE')
  L_VARIABLE=[]
  if VARIABLE:
    if 'VARIABLE' not in keypoint:
      keypoint.append('VARIABLE')
    st1=''
    for B in VARIABLE:
      s=list(B.attrib.values())
      st1 +=str(s[0])+' '+str(s[1])+" "
    L_VARIABLE.append(st1)
    va.append(L_VARIABLE)
  else:
    if 'VARIABLE' not in keypoint:
      keypoint.append('VARIABLE')
    va.append('')


  RBC=r.findall('RULE_BASED_CALENDAR')
  L_RBC=[]
  if RBC:
    if 'RBC' not in keypoint:
      keypoint.append('RBC')
    st=''
    for a in RBC:
      s=list(a.attrib.values())
      st +=" ".join(s)+"     "
    L_RBC.append(st)
    va.append(L_RBC)
  else:
    if 'RBC' not in keypoint:
      keypoint.append('RBC')
    va.append('')    

  CONTROL=r.findall('CONTROL')
  L_CONTROL=[]
  if CONTROL:
    if 'CONTROL' not in keypoint:
      keypoint.append('CONTROL')
    st=''
    for a in CONTROL:
      s=list(a.attrib.values())
      st +=str(s[0])+' '+str(s[1])+" "
    L_CONTROL.append(st)
    va.append(L_CONTROL)
  else:
    if 'CONTROL' not in keypoint:
      keypoint.append('CONTROL')
    va.append('')
      
  INCOND=r.findall('INCOND')
  L_INCOND=[]
  if INCOND:
    if 'INCOND' not in keypoint:
      keypoint.append('INCOND')
    st=''
    for a in INCOND:
      s=list(a.attrib.values())
      st +=str(s[0])+' '+str(s[1])+" "+str(s[2])+" "
    L_INCOND.append(st)
    va.append(L_INCOND)
  else:
    if 'INCOND' not in keypoint:
      keypoint.append('INCOND')
    va.append('')
      
      
  OUTCOND=r.findall('OUTCOND')
  L_OUT=[]
  if OUTCOND:
    if 'OUTCOND' not in keypoint:
      keypoint.append('OUTCOND')
    st=''
    for a in OUTCOND:
      s=list(a.attrib.values())
      st +=str(s[0])+' '+str(s[1])+" "+str(s[2])+" "
    L_OUT.append(st)
    va.append(L_OUT) 
  else:
    if 'OUTCOND' not in keypoint:
      keypoint.append('OUTCOND')
    va.append('')

  ON=r.findall('ON')
  L_ON=[]
  if ON:
    if 'ON' not in keypoint:
      keypoint.append('ON')
    key=['CODE']
    st=''
    for p in ON:
      k=list(p.attrib.keys())
      v=list(p.attrib.values())
      for q in key:
        if q in k:
          st +=str(v[k.index(q)])+" "
          DOACTION=p.findall('DOACTION')
          if DOACTION:
            for aB in DOACTION:
              l=list(aB.attrib.values())
              st +=" ".join(l)+" "
          DOCOND=p.findall('DOCOND')
          if DOCOND:
            for aB in DOCOND:
              l=list(aB.attrib.values())
              st +=" ".join(l)+" "
          DOEMAIL=p.findall('DOMAIL')
          if DOEMAIL:
            for aB in DOEMAIL:
              l=list(aB.attrib.values())
              st +=" ".join(l)+" "  
          DOSHOUT=p.findall('DOSHOUT')
          if DOSHOUT:
            for aB in DOSHOUT:
              l=list(aB.attrib.values())
              st +=" ".join(l)+" "  
    L_ON.append(st)
    va.append(L_ON)           
  else:
    if 'ON' not in keypoint:
      keypoint.append('ON')
    va.append('')

  SHOUT=r.findall('SHOUT')
  L_SHOUT=[]
  if SHOUT:
    if 'NOTFY BEFOR(0)/AFTER' not in keypoint:
      keypoint.append('NOTFY BEFOR(0)/AFTER')
    st=''
    for a in SHOUT:
      s=list(a.attrib.values())
      st +=" ".join(s)+" "
    L_SHOUT.append(st)
    va.append(L_SHOUT) 
  else:
    if 'NOTFY BEFOR(0)/AFTER' not in keypoint:
      keypoint.append('NOTFY BEFOR(0)/AFTER')
    va.append('')
  vaa.append(va)

with open(m+"_"+e+"_FOLDER.csv", "w", newline='') as f:
    writer = csv.writer(f, delimiter=',')
    writer.writerow(keypoint) # write the header
    # write the actual content line by line
    for l in vaa:
        writer.writerow(l)    
print("succuss_FOLDER")

```
